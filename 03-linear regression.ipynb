```python
# 03 线性回归与代价函数

## 一、线性回归模型

### 模型表示
$$f_{w,b}(x) = wx + b$$

| 符号 | 名称 | 含义 |
|------|------|------|
| $f_{w,b}(x)$ | 假设函数（hypothesis） | 模型的预测输出 |
| $w$ | 权重（weight）/斜率 | 控制线的倾斜程度 |
| $b$ | 偏置（bias）/截距 | 控制线的上下位置 |
| $x$ | 输入特征 | 如房屋面积 |
| $y$ | 真实值（label） | 实际房价 |

&gt; **为什么用 $f$ 不用 $y$？**
&gt; $y$ 是真实值，$f(x)$ 是预测值，两者必须区分。

&gt; **打个比方**：$w$ 就像方向盘的灵敏度，$b$ 就像方向盘的初始偏移。调这两个参数，就是在找一条最贴合数据的直线。

---

```
