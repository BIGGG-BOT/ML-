```python
# 三、线性回归模型

## 模型表示

$$f_{w,b}(x) = wx + b$$

- **$f$**：假设函数（hypothesis），模型的预测输出
- **$w$**：权重（weight）/斜率，控制线的倾斜程度
- **$b$**：偏置（bias）/截距，控制线的上下位置
- **$x$**：输入特征（如房屋面积）

&gt; **为什么用 $f$ 不用 $y$？**
&gt; 因为 $y$ 是真实值，$f(x)$ 是预测值，两者要区分。

## 打比方

$w$ 就像**方向盘的灵敏度**，$b$ 就像**方向盘的初始偏移**。
调这两个参数，就是在找一条最贴合数据的直线。

---

```
