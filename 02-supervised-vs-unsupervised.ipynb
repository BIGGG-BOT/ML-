# 02 监督学习 vs 无监督学习

## 监督学习（Supervised Learning）
给定输入 x（特征）和正确答案 y（标签），学习 x→y 的映射。  
就像学生做题，有标准答案可以对。

| 类型 | 预测目标 | 例子 |
|------|---------|------|
| **分类** | 离散标签 | 垃圾邮件 / 正常邮件 |
| **回归** | 连续值 | 房价、温度 |

## 无监督学习（Unsupervised Learning）
只给 x，不给 y，让机器自己找结构。  
就像把一堆照片按人脸自动分组，没人告诉你哪张是谁。

| 类型 | 做什么 | 算法 |
|------|--------|------|
| **聚类** | 自动分组 | K-Means、层次聚类 |
| **降维** | 压缩信息 | PCA、t-SNE |
| **异常检测** | 找出"不正常"的数据点 | — |

## 核心区别
- 监督学习：**有标签**（知道答案）
- 无监督学习：**无标签**（自己摸索）
