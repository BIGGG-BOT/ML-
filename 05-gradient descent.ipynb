# 梯度下降（Gradient Descent）

## 一、本质
梯度下降就是系统性地找到最好的 w 和 b，让代价函数 J(w,b) 最小。可以适用与所有函数。
可能有多个最小值
## 二、打比方：盲人下山
想象一个盲人站在山顶，想走到山底：
1. 他用脚探一探脚下哪个方向最陡（算导数）
2. 然后往反方向迈一步
3. 重复直到脚底平坦（导数≈0，到谷底了）

## 三、更新规则
w := w - α ∂J/∂w (让w参数减去学习速率乘以代价函数的导数。（以w为参数求导，b为常数）)
b := b - α ∂J/∂b

关键：同时更新——先算好两个新值，再一起换

## 四、学习率 α
| 情况 | 结果 | 打比方 |
|------|------|--------|
| 太小 | 收敛极慢 | 蜗牛爬 |
| 合适 | 稳步收敛 | 正常步伐 |
| 太大 | 震荡或发散 | 跨过头滚下山 |

## 五、局部最小值 vs 全局最小值
线性回归的代价函数是凸函数（碗状），只有一个谷底，不担心局部最小值。
