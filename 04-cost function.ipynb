```python
## 代价函数

量化"模型预测值和真实值的差距"，差距越大，代价越高。

$$J(w,b) = \frac{1}{2m} \sum_{i=1}^{m} \left( f_{w,b}\left(x^{(i)}\right) - y^{(i)} \right)^2$$

| 符号 | 含义 |
|------|------|
| $m$ | 训练集样本数量 |
| $x^{(i)}$ | 第 $i$ 个样本的输入 |
| $y^{(i)}$ | 第 $i$ 个样本的真实值 |
| $f_{w,b}(x^{(i)})$ | 第 $i$ 个样本的预测值 |

**前面为什么乘 $\frac{1}{2}$？**

纯数学技巧：求导时和平方的 2 抵消，简化计算。

---

## 平方误差 vs 绝对值误差

| 维度 | 平方误差（MSE） | 绝对值误差（MAE） |
|------|----------------|------------------|
| 可导性 | ✅ 处处可导，梯度下降能跑 | ❌ 在误差=0 处不可导（尖点） |
| 直观比喻 | 光滑的山坡，可以滚到谷底 | 带台阶的楼梯，走到边缘会卡住 |
| 统计理论 | 对应高斯分布的极大似然估计 | 对应拉普拉斯分布 |
| 对异常值 | 放大惩罚（$50^2=2500$ vs $10^2=100$） | 一视同仁 |

**核心原因：可导性**

平方误差是光滑函数，梯度下降能沿着导数方向一步步逼近最优；绝对值有"尖点"。
```
