```python
# 四、代价函数（Cost Function）

## 本质

量化"模型预测值和真实值的差距"，差距越大，代价越高。

## 公式

$$J(w,b) = \frac{1}{2m}\sum_{i=1}^{m}(f_{w,b}(x^{(i)}) - y^{(i)})^2$$

## 为什么用平方误差，不用绝对值？

| 维度 | 平方误差（MSE） | 绝对值误差（MAE） |
|------|----------------|------------------|
| 可导性 | ✅ 处处可导 | ❌ 尖点不可导 |
| 打比方 | 光滑山坡 | 带台阶楼梯 |
| 统计理论 | 高斯分布MLE | 拉普拉斯分布 |
| 异常值 | 放大惩罚 | 一视同仁 |

&gt; **核心原因：可导性**
```
